```python
cat={}
category={}
for i in data['Category'].unique():
    free_count=0
    paid_count=0
    for j in data.values:
        if(j[1]==i):
            if(j[6]=="Free"):
                free_count+=1
            else:
                paid_count+=1
    cat[i]=[free_count,paid_count]
    category[i]={"Free":free_count,"Paid":paid_count}
            
```

### Calculating all of the Dif Categories of Every App

```python
for i in category:
    print(i,category[i])
```

```python
for j in cat:
    print(j,cat[j])
```

# GroupBy Function

```python
data = pd.read_csv("googleplaystore.csv")
```

```python
df = data.groupby('Category').mean()['Rating'].sort_values(ascending=False)
df
```

```python
df = data.groupby('Category').count()['Type'].sort_values(ascending=False)
df
```
